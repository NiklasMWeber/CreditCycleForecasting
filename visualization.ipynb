```python
# Delete unemployment from data, to make signature usable. 
# Otherwise singature will be to long and number of betas higher than number of observations.
X = np.delete(X,0,1)
```

```python
# Standardize Data
scaler = pre.StandardScaler()
X_scaled = scaler.fit_transform(X) #subtract mean and divide by std
max_abs_scaler = pre.MaxAbsScaler()
Y_scaled = max_abs_scaler.fit_transform(Y-np.mean(Y))# subtract mean and min-max-transform into range [-1,1]
```

```python
# Construct 3 year rolling windows:
reg_data = []
predictors = []
predictors_for_Signature = []
for i in range(3,len(year)):
    predictors.append(X_scaled[(i-3):i,:].reshape(-1))  
    predictors_for_Signature.append(X_scaled[(i-3):i,:])
    
reg_data = [np.array(predictors),Y_scaled[3:len(year)]]
```

```python
# Split in train and test sample

#randomNumbers = np.random.randn(reg_data[0].shape[0], reg_data[0].shape[1])
#X_train, X_test, Y_train, Y_test = train_test_split( reg_data[0],reg_data[1], test_size = 0.2)

train_idx, test_idx, Y_train, Y_test = train_test_split( range(reg_data[0].shape[0]),reg_data[1], test_size = 0.15)
X_train = reg_data[0][train_idx,:] 
#randomNumbers = np.random.randn(reg_data[0].shape[0], reg_data[0].shape[1])
#X_train = randomNumbers
X_test = reg_data[0][test_idx,:]
print('Number of predictors/betas to be estimated:', X_train.shape[1])
print('Number of Observation tuples for estimation:', X_train.shape[0])
#print(X_test_idx)
#print(X_test)
```

```python
# Train and fit ridge regression using 3 year wrolling window to predict following year PD
ridge_reg = Ridge(alpha = 1, fit_intercept=True)
ridge_reg.fit(X_train,Y_train)

predict_test = ridge_reg.predict(X_test)
predict_train = ridge_reg.predict(X_train)
```

```python
# Plot
fig = plt.figure()
fig1 = fig.add_subplot(2,1,1)
xAxes = range(1,np.shape(predict_test)[0]+1)
fig1.plot(xAxes, Y_test, label = 'true')
fig1.plot(xAxes, predict_test, label = 'predict')
fig1.legend()
fig1.set_title('Testing Sample')
print('R on Test Sample = ', ridge_reg.score(X_test, Y_test))

fig2 = fig.add_subplot(2,1,2)
xAxes = range(1,np.shape(predict_train)[0]+1)
fig2.plot(xAxes, Y_train, label = 'true')
fig2.plot(xAxes, predict_train, label = 'predict')
fig2.legend()
fig2.set_title('Training Sample')
print('R on Training Sample = ', ridge_reg.score(X_train, Y_train))
```

```python
# Calculate singature of the 3-year rolling window path:
depth = 2
path_dimension = predictors_for_Signature[0].shape[1]
print('Depth of Signature set to:', depth)
print('Path dimension is:',path_dimension)
print('Length of Signature will be:', ts.sigdim(path_dimension,depth))
signatures = []
for path in predictors_for_Signature:
    signatures.append(ts.stream2sig(path,depth))
signatures = np.array(signatures)
```

```python
# Use same training test split for regression on signatures:
# Y_train and Y_test stay the same
X_train = signatures[train_idx,:]
X_test = signatures[test_idx,:]

print('Number of predictors/betas to be estimated:', X_train.shape[1])
print('Number of Observation tuples for estimation:', X_train.shape[0])
```

```python
# Train and fit ridge regression using signatures of multi dimensional 3 year path to predict following year PD
ridge_reg = Ridge(alpha = 0.1, fit_intercept=True)
ridge_reg.fit(X_train,Y_train)

predict_test = ridge_reg.predict(X_test)
predict_train = ridge_reg.predict(X_train)
```
